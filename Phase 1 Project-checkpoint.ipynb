## Data Cleaning
Now that we have inspected the data, the next step ideally would be to clean the data.
Some tables have columns with a high percentage of null values that need to be dealt with, whereas some have a low percentage where the values can either be replaced or dropped.
Furthermore, some dataframes also have duplicates that need to be dropped so that we can be able to proceed to data analysis and visualization.
The aim here therefore is:

        >> To deal with null values accordingly
        >> To drop duplicates
The completion of this stage would be very fundamental to hasten the next steps where we get to manipulate our data and derive conclusions from the data given.      

### Table 1: bom_movie_gross
This table contains missing values that are less than 50%.
The columns affected are relatively important. The median would be the best fit to replace this missing values.
Remember, the foreign gross column is an object! 
It therefore needs to be transformed to an integer or float in order to perform mathematical operations smoothly.

```python
# Find the non-null values in the foreign gross column
# Convert the column to integer/float
# Replace the null values with the median
bom_movie_gross_clean = bom_movie_gross[bom_movie_gross['foreign_gross'].notnull()]
f_gross_median = bom_movie_gross_clean['foreign_gross'].str.replace(',','').astype(float).median().round()
f_gross_median
```

```python
# Replace the missing values
bom_movie_gross['foreign_gross'].fillna(f_gross_median, inplace= True)
bom_movie_gross.head()
```

```python
# Do the same for domestic gross column
bom_movie_gross_clean = bom_movie_gross[bom_movie_gross['domestic_gross'].notnull()]
bom_movie_gross_clean
```

```python
d_gross_median = bom_movie_gross_clean.domestic_gross.median().round()
d_gross_median
```

```python
# Replace the missing values in the domestic gross column
bom_movie_gross['domestic_gross'].fillna(d_gross_median, inplace= True)
bom_movie_gross.head()
```

```python
# Define a function to fill null values
def fill_null(data, col):
    df = data[col].fillna('not_stated', inplace = True)
    
    return df
```

```python
#A function to help us plot bar graphs
def bar_graph(data, var_x, var_y):
    plt.figure(figsize = (14,10))
    sns.set_theme(style="darkgrid")
    ax = sns.barplot(x= data[var_x],
                     y= data[var_y],
                     data= data, palette = 'flare')
    ax.set_xticklabels(ax.get_xticklabels(),rotation = 90, fontsize= 15)
    ax.set_xlabel(f"{var_x.title()}", fontsize= 20)
    ax.set_ylabel(f"{var_y.title()}", fontsize= 20)
    ax.set_title(f"{var_x.title()} vs {var_y.title()}", fontsize = 20)
    plt.savefig(f"{var_x.title()} vs {var_y.title()}")
    plt.show()
```

### Question 4: What is the duration of majority of the movies?

```python
movies_joined.head()
```

```python
# Plot a histogram to visualize this
# Define a function to plot future histograms

def plot_hist(data, var_x):
    plt.figure(figsize = (14,10))
    sns.set_theme(style="darkgrid")
    ax = sns.histplot(x= data[var_x],
                     data= data, palette = 'flare')
    mean = data[var_x].mean()
    ax.axvline(mean, color = 'red')
    ax.set_xlabel(f"{var_x.title()}", fontsize= 20)
    ax.set_title(f"{var_x.title()} Distribution", fontsize = 20)
    plt.savefig(f"{var_x.title()}",dpi=400)
    plt.show()
    
    print("The mean of {} is: {}". format(var_x, mean))
```

```python
plot_hist(movies_joined, 'runtime_minutes')
```

From the diagram above, a majority of the movies have a runtime centered between 80 and 125 minutes with the average value being 104 minutes.

The movie duration is important as the length of the movie influences the audience's attention.
Very long movies might end up boring a person and conversely, short movies may just hasten the storyline hence not fun to watch. Keep in mind that this might not always necessarily be the case depending on the target audience's preference or taste in movies. 
While this is the case, we may arrive at a general conclusion that there is no ideal length of movie runtimes. But, it would be great to base and kick off with the most frequent runtime followed by most movies as depicted above.

```python
# Let's look at how the rating differs by runtime
rating_vs_runtime_1 = movies_joined_ratings[movies_joined_ratings['runtime_minutes'] < 80].sort_values (by = 'averagerating' ,ascending = False).head()
rating_vs_runtime_1

# Add a new column to the dataframe for less than 80 mins runtime
rating_vs_runtime_1['less_80_mins'] = '< 80'
rating_vs_runtime_1
```

```python
# Do the same for more than 80 mins
rating_vs_runtime = movies_joined_ratings[movies_joined_ratings['runtime_minutes'] > 80].sort_values (by = 'averagerating' ,ascending = False).head()
rating_vs_runtime

# Add a new column to the dataframe for less than 80 mins runtime
rating_vs_runtime['more_80_mins'] = '> 80'
rating_vs_runtime
```

```python
# Plot the figure
fig, ax1 = plt.subplots(figsize = (10, 8))
sns.set_theme(style="darkgrid")
ax1  = sns.barplot(x= rating_vs_runtime_1['less_80_mins'],
                   y= rating_vs_runtime_1['numvotes'],
                  data= rating_vs_runtime_1, palette = 'flare')
ax1.set_xlabel("< 80", fontsize= 20)
ax1.set_ylabel("Number of votes", fontsize= 20)
ax1.set_title("Comparison of number of votes to the runtime length", fontsize = 15)
plt.savefig('less_80_mins.png',dpi=400)

plt.show()
```

```python
fig, ax2 = plt.subplots(figsize = (10,8))
ax2  = sns.barplot(x= rating_vs_runtime['more_80_mins'],
                   y= rating_vs_runtime['numvotes'],
                  data= rating_vs_runtime, palette = 'flare')
ax2.set_xlabel("> 80", fontsize= 20)
ax2.set_ylabel("Number of votes", fontsize= 20)
ax2.set_title("Comparison of number of votes to the runtime length", fontsize = 15)
plt.savefig('more_80_mins.png',dpi=400)

plt.show()
```

As derived from the two figures displayed above, more number of votes have been cast for movies with a runtime of above 80 minutes. Slightly above 600,000 people created time to vote for movies with this runtime length (> 80).

On the other hand, very few people showed interest in voting for movies with a lower than 80 minutes runtime length. Close to 35 people voted for these flicks. 
This is a huge contrast and how the film audience reacts is fundamental to keep track of the productions. If a movie receives a high number of votes from people, it results in the movie making more profit.
